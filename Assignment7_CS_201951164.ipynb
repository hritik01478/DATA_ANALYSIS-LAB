```python
df=pd.read_csv('melb_data.csv')
df.head()
```

```python
data=df[['Price','Landsize']]
data.head()
```

```python
data.shape
```

```python
data.isna().sum()
```

```python
data1="Landsize"
data2="Price"
hb=plt.hexbin(data[data1], 
           data[data2], 
            gridsize = 100,
           cmap ='Greens') 

plt.xlabel(data1)
plt.ylabel(data2)
plt.xlim(0,4000)
plt.ylim(0,1500000)
cb = plt.colorbar(hb) 
```
